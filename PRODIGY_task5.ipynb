## Multiple Variable  Non-Graphical Summary

```python
numeric_columns = ['Severity', 'Start_Lat', 'Start_Lng', 'Distance(mi)',
                   'Temperature(F)', 'Visibility(mi)', 'Wind_Speed(mph)']

# Correlation matrix
correlation_matrix = df[numeric_columns].corr()
print(correlation_matrix)
```

```python
df[numeric_columns].agg(['min', 'max', 'mean', 'median', 'std', 'skew'])
```

```python
pd.crosstab(df['Severity'], df['Sunrise_Sunset'])
```

```python
boolean_columns = ['Amenity', 'Crossing', 'Junction', 'Traffic_Signal']

for col in boolean_columns:
    print(f"Crosstab for Severity vs {col}")
    print(pd.crosstab(df['Severity'], df[col]))
    print("\n")
```
